```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import pickle
```

```python
data = pd.read_csv("C:\kaggle datasets\diabetes with more features\dataset\diabetes_binary_health_indicators_BRFSS2015.csv")
```

```python
data.sample(10)
```

```python
data.columns
```

```python
drop_col = ['Education','Income']
data.drop(data[drop_col],axis=1 ,inplace=True)
```

```python
data.head()
```

```python
data.shape
```

```python
y = data['Diabetes_binary']
x = data.iloc[: ,1:]
```

```python
sc = StandardScaler()
x = sc.fit_transform(x)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.30,random_state=1)
```

```python
rf = RandomForestClassifier(n_estimators=150,)
rf.fit(x_train,y_train)
```

```python
accuracy = rf.predict(x_test)
pred = accuracy_score(accuracy,y_test)
print(pred)
```

```python
input = (1,1,1,40,1,0,0,0,0,1,0,1,0,5,18,15,1,0,9)
inp_arr = np.asarray(input)
inp_arr_reshaped  = inp_arr.reshape(1,-1)
std_data = sc.transform(inp_arr_reshaped)
prediction = rf.predict(std_data)
print(prediction)
```
